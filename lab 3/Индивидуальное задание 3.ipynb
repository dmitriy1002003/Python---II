## Индивидуальное задание 3
Вычислите определенный интеграл для функции:
f(x) = x3 − 2x2 + x
на отрезке [−1, 2].


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Определение функции f(x) = x^3 - 2x^2 + x
def f(x):
    return x**3 - 2*x**2 + x

# Границы интегрирования
a, b = -1, 2

# Вычисление определенного интеграла
integral_value, _ = quad(f, a, b)

# Создание массива точек для графика
x = np.linspace(-1.5, 2.5, 400)
y = f(x)

# Построение графика
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b', linewidth=2, label=r'$f(x) = x^3 - 2x^2 + x$')

# Закраска области под графиком
x_fill = np.linspace(a, b, 200)
y_fill = f(x_fill)
plt.fill_between(x_fill, y_fill, color='blue', alpha=0.3, label=f'Интеграл = {integral_value:.2f}')


plt.axhline(0, color='black', linewidth=1)  # Ось X
plt.axvline(0, color='black', linewidth=1)  # Ось Y
plt.title("Визуализация определенного интеграла", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Отображение графика
plt.show()
```
