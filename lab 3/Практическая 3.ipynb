```python
import numpy as np
import matplotlib.pyplot as plt

# Интервал значений x
x = np.linspace(-10, 10, 50)

# Квадратичная зависимость
y = [i**2 for i in x]

# Построение графика
plt.title("График функции y = x^2")  # Заголовок
plt.xlabel("x")  # Подпись оси x
plt.ylabel("y")  # Подпись оси y
plt.grid()  # Включение сетки
plt.plot(x, y)  # Построение графика

# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-3, 3, 50)

y1 = x  # Линейная функция
y2 = x**2  # Квадратичная функция
y3 = x**3  # Кубическая функция

# Построение графика
plt.figure(figsize=(7, 5))
plt.plot(x, y1, 'b--', linewidth=2, label=r'$y = x$')   # Синяя пунктирная линия
plt.plot(x, y2, 'g-.', linewidth=2, label=r'$y = x^2$') # Зеленая штрих-пунктирная линия
plt.plot(x, y3, 'r-', linewidth=2, label=r'$y = x^3$')  # Красная сплошная линия

plt.title("Графики функций $y = x$, $y = x^2$, $y = x^3$", fontsize=14)  # Заголовок
plt.xlabel("x", fontsize=12)  # Подпись оси x
plt.ylabel("y", fontsize=12)  # Подпись оси y
plt.grid()  # Включение сетки
plt.legend()  # Добавление легенды
plt.ylim(-10, 10)  # Установил лимиты по y для лучшей видимости

# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация случайных точек
np.random.seed(42)
x = np.random.rand(50) * 10  # Координаты x 
y = np.random.rand(50) * 10  # Координаты y 
sizes = y * 10  # Размер точек зависит от y
colors = x  # Цвет точек зависит от x

# Построение диаграммы рассеяния
plt.figure(figsize=(7, 5))
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7, edgecolors='black')


plt.title("Диаграмма рассеяния с цветом и размером точек", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)

# Отображение графика
plt.show()
```

```python
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Вычисление среднего значения
mean_value = np.mean(data)

# Построение гистограммы
plt.figure(figsize=(7, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Линия среднего значения
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Среднее значение')


plt.title("Гистограмма нормального распределения", fontsize=14)
plt.xlabel("Значение", fontsize=12)
plt.ylabel("Частота", fontsize=12)
plt.legend()

# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создание сетки значений x и y
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Вычисление значений z
Z = np.sin(np.sqrt(X**2 + Y**2))

# Построение 3D-графика
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')


ax.set_title(r"3D-график функции $z = \sin(\sqrt{x^2 + y^2})$", fontsize=14)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

# Отображение графика
plt.show()
```

```python
x = np.linspace(-10, 10, 100)

# Создание фигуры и подграфиков (2x2)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Линейный график y = x
axes[0, 0].plot(x, x, 'b')
axes[0, 0].set_title(r'Линейный график $y = x$')

# Парабола y = x^2
axes[0, 1].plot(x, x**2, 'g')
axes[0, 1].set_title(r'Парабола $y = x^2$')

# Синус y = sin(x)
axes[1, 0].plot(x, np.sin(x), 'r')
axes[1, 0].set_title(r'Синус $y = \sin(x)$')

# Косинус y = cos(x)
axes[1, 1].plot(x, np.cos(x), 'm')
axes[1, 1].set_title(r'Косинус $y = \cos(x)$')


for ax in axes.flat:
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True, linestyle="--", alpha=0.6)

# Оптимизация расположения
plt.tight_layout()

# Отображение графика
plt.show()
```
