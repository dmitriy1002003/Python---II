```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Определение функции f(x, y) = 1 / (1 + x^2 + y^2)
def f(x, y):
    return 1 / (1 + x**2 + y**2)

# Создание сетки значений x, y в диапазоне [-3, 3]
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Построение 3D-графика поверхности
fig = plt.figure(figsize=(10, 5))

# График поверхности (plot_surface)
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax1.set_title("Плотная поверхность")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("f(x, y)")

# Каркасный график (plot_wireframe)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(X, Y, Z, color
                ='black', alpha=0.6)
ax2.set_title("Каркасная модель")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("f(x, y)")

# Отображение графиков
plt.show()
```
