```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
industries = ['IT', 'Производство', 'Туризм', 'Финансы', 'Транспорт']
gdp_share = [15, 25, 10, 30, 20]

# Вычисление среднего значения
mean_value = np.mean(gdp_share)

# Построение столбчатой диаграммы
plt.figure(figsize=(8, 5))
bars = plt.bar(industries, gdp_share, color=['blue', 'green', 'orange', 'red', 'purple'])

# Подписи значений сверху столбцов
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height}%', ha='center', va='bottom', fontsize=12)

# Добавление горизонтальной линии среднего значения
plt.axhline(mean_value, color='black', linestyle='dashed', linewidth=2, label=f'Среднее: {mean_value:.1f}%')


plt.title("Вклад разных отраслей в ВВП", fontsize=14)
plt.xlabel("Отрасли", fontsize=12)
plt.ylabel("Доля в ВВП (%)", fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle="--", alpha=0.6)

# Отображение графика
plt.show()
```
