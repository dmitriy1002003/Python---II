## Индивидуальное задание 1
Создайте CSV-файл sales.csv со следующими данными:
Дата,Продажи
2024-07-01,120
2024-07-02,150
2024-07-03,170
2024-07-04,160
2024-07-05,180
Прочитайте данные, установите DatetimeIndex , найдите скользящее среднее
( rolling(2).mean() ) и постройте два графика: реальных продаж и их сглаженного
значения.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Шаг 1: Считываем данные из CSV
# Предположим, файл sales.csv имеет структуру:
# Дата,Продажи
# 2023-07-01,120
# 2023-07-02,150
# 2023-07-03,170
# ...
df = pd.read_csv('sales.csv', parse_dates=['Дата'], index_col='Дата')

# Шаг 2: Убеждаемся, что индекс стал DatetimeIndex
print(df.index)  # Должен отобразиться DatetimeIndex

# Шаг 3: Вычисляем скользящее среднее с окном 2
df['СкользящееСреднее'] = df['Продажи'].rolling(window=2).mean()

# Шаг 4: Строим график реальных продаж и скользящего среднего
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['Продажи'], marker='o', label='Реальные продажи')
plt.plot(df.index, df['СкользящееСреднее'], marker='o', label='Скользящее среднее (окно=2)')

# Шаг 5: Оформляем график
plt.title("Анализ продаж в интернет-магазине")
plt.xlabel("Дата")
plt.ylabel("Продажи")
plt.legend()
plt.grid(True)
plt.show()
```
